**Problem 3:** Create a bar chart using Matplotlib to visualize the following data: Monthly sales for a store (January to May). Use appropriate labels and titles for the chart.

````python

months = ("jan", "feb", "mar", "apr", "may")
salesData = {
    'Store1': (18734.35, 34833.43, 23532.98, 39328.15, 26348.85),
    'Store2': (20321.79, 36386.83, 21437.50, 34342.43, 28567.37),
}

x = np.arange(len(months))  # Use months for x-axis positions
width = 0.25
multiplier = 0

fig, ax = plt.subplots()

for attribute, measurement in salesData.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    ax.bar_label(rects, padding=3)
    multiplier += 1

ax.set_ylabel('Sales Amount')
ax.set_title('Monthly Sales by Store')
ax.set_xticks(x + width / 2)  # Center the bars on the x-axis
ax.set_xticklabels(months)  # Set x-axis labels
ax.legend(loc='upper left', title="Stores")

ax.set_ylim(0, max([max(measurement) for measurement in salesData.values()]) + 1000)

plt
````

**Problem 10:** Given a Pandas DataFrame containing information about students, add a new column 'Grade' based on the following criteria:
* If 'Score' >= 90, Grade is 'A'
* If 'Score' >= 80, Grade is 'B'
* If 'Score' >= 70, Grade is 'C'
* If 'Score' < 70, Grade is 'D'

````python


data = {'Name': ['Aria', 'Ali', 'Saman', 'Ahmad'],
        'Score': [95, 78, 67, 88]}
df = pd.DataFrame(data)


conditions = [
    df['Score'] >= 90,
    (df['Score'] >= 80) & (df['Score'] < 90),
    (df['Score'] >= 70) & (df['Score'] < 80),
    df['Score'] < 70
]

grades = ['A', 'B', 'C', 'D']


df['Grade'] = np.select(conditions, grades, default='F')


print(df)
````

**Problem 11:** You are given a dataset containing monthly sales data for three products (A, B, and C) over a two-year period. Your task is to perform various data analysis tasks using NumPy, Pandas, and Matplotlib.


*   Dataset:

```
import pandas as pd

data = {
    'Month': pd.date_range(start='2021-01-01', periods=24, freq='M'),
    'Product A Sales': [500, 480, 600, 750, 900, 850, 920, 1100, 1300, 1350, 1500, 1450, 1550, 1600, 1650, 1600, 1500, 1400, 1600, 1700, 1800, 1750, 1850, 1900],
    'Product B Sales': [300, 320, 400, 450, 500, 580, 700, 750, 820, 900, 950, 980, 1050, 1100, 1150, 1200, 1250, 1300, 1350, 1400, 1500, 1600, 1550, 1700],
    'Product C Sales': [200, 210, 250, 280, 320, 350, 380, 400, 420, 440, 460, 480, 500, 520, 540, 560, 580, 600, 620, 640, 660, 680, 700, 720]
}

sales_df = pd.DataFrame(data)
```
*  Tasks:
1. Calculate and display the total sales for each product over the two-year period.

2. Calculate and display the average monthly sales for each product.

3. Find the month with the highest sales for each product, and display the product and the sales value for that month.

4. Calculate and display the percentage change in sales for each product from January to December in the second year (2022).

5. Create a line chart using Matplotlib to visualize the monthly sales data for each product over the two-year period. Label the axes and provide a title for the chart.

6. Calculate the correlation between Product A and Product B sales. Is there a strong correlation between these two products? Display the correlation coefficient.


````python

data = {
    'Month': pd.date_range(start='2021-01-01', periods=24, freq='M'),
    'Product A Sales': [500, 480, 600, 750, 900, 850, 920, 1100, 1300, 1350, 1500, 1450, 1550, 1600, 1650, 1600, 1500, 1400, 1600, 1700, 1800, 1750, 1850, 1900],
    'Product B Sales': [300, 320, 400, 450, 500, 580, 700, 750, 820, 900, 950, 980, 1050, 1100, 1150, 1200, 1250, 1300, 1350, 1400, 1500, 1600, 1550, 1700],
    'Product C Sales': [200, 210, 250, 280, 320, 350, 380, 400, 420, 440, 460, 480, 500, 520, 540, 560, 580, 600, 620, 640, 660, 680, 700, 720]
}

sales_df = pd.DataFrame(data)

# Task 1: Calculate and display the total sales for each product over the two-year period.
total_sales = sales_df.sum()
print("Total Sales for Each Product:")
print(total_sales[1:])

# Task 2: Calculate and display the average monthly sales for each product.
average_monthly_sales = sales_df.mean()
print("\nAverage Monthly Sales for Each Product:")
print(average_monthly_sales[1:])

# Task 3: Find the month with the highest sales for each product.
max_sales_month = sales_df[['Product A Sales', 'Product B Sales', 'Product C Sales']].idxmax()
max_sales = sales_df[['Product A Sales', 'Product B Sales', 'Product C Sales']].max()
print("\nMonth with Highest Sales for Each Product:")
for product in ['Product A', 'Product B', 'Product C']:
    product_col = product + ' Sales'
    max_month_idx = max_sales_month[product_col]
    max_month = sales_df['Month'].loc[max_month_idx].strftime('%B %Y')
    max_value = max_sales[product_col]
    print(f"{product}: {max_month}, Sales: {max_value}")

# Task 4: Calculate and display the percentage change in sales from January to December in the second year (2022).
january_sales = sales_df.loc[sales_df['Month'] == '2022-01-31']
december_sales = sales_df.loc[sales_df['Month'] == '2022-12-31']
percentage_change = (december_sales.iloc[:, 1:] - january_sales.iloc[:, 1:]) / january_sales.iloc[:, 1:] * 100
print("\nPercentage Change in Sales from January to December (2022):")
print(percentage_change)

# Task 5: Create a line chart to visualize the monthly sales data for each product over the two-year period.
plt.figure(figsize=(10, 6))
for product in ['Product A Sales', 'Product B Sales', 'Product C Sales']:
    plt.plot(sales_df['Month'], sales_df[product], label=product)

plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Monthly Sales Data Over Two Years')
plt.legend()
plt.grid(True)
plt.show()

# Task 6: Calculate the correlation between Product A and Product B sales.
correlation = sales_df['Product A Sales'].corr(sales_df['Product B Sales'])
print("\nCorrelation between Product A and Product B Sales:")
print(f"Correlation Coefficient: {correlation}")
````
